```python
%matplotlib  
%matplotlib inline

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_color = cv.imread('arandelas.jpg')
imgRGB = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)

plt.figure(1)
plt.imshow(imgRGB)
plt.show()
```

```python
img_HSV = cv.cvtColor(img_color, cv.COLOR_BGR2HSV)
plt.figure(5)
plt.imshow(img_HSV)

img_muestra_hsv = img_HSV[230:280,230:280,:]

color_mean = cv.mean(img_muestra_hsv)
```

# Amarillo

```python
# Tomamos una muestra del morron amarillo

#img_muestra = imgRGB[100:150,325:375,:]
img_muestra = imgRGB[1060:1100,1800:1850,:]
plt.figure(2)
plt.imshow(img_muestra)
plt.show()

color_mean, color_std = cv.meanStdDev(img_muestra)
    
print(color_mean.flatten().tolist())
print(color_std.flatten().tolist())
```

```python
# Pintemos de blanco los puntos que estén cerca de color_mean en un radio menor que 3sigmas

mask = cv.inRange(imgRGB, color_mean-color_std*8,  color_mean+color_std*8)
img_segmentada = cv.bitwise_and(imgRGB, imgRGB, mask=mask)

#plt.imshow(imgRGB)
#plt.show()

plt.figure(3)
plt.imshow(img_segmentada)
plt.show()
```
